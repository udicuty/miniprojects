```python
#Dataframe sorted by 
def SE(x):
    n=np.sqrt(x.count())
    if n==0:
        return None
    ans=np.std(x)
    return(ans)
a=df.groupby('ZIPCODE').agg({"CAMIS":lambda x: x.nunique(),"SCORE":np.mean,"SD":np.std,"SE":SE})
a
#b=a.CAMIS.nunique().sort_values(ascending=False).head(100)
```

```python
topzipcodes=a[a.CAMIS>100].sort_values('SCORE',ascending=False)
topzipcodes["ZIPCODE"]=topzipcodes.index
topzipcodes.shape
topzipcodes.head(5)
#topzipcodes[1]

#a=pd.DataFrame({a:topzipcodes.ZIPCODE,b:topzipcodes.SD})
topzipcodes= topzipcodes[['ZIPCODE', 'SCORE', 'SE', 'CAMIS']]
topzipcodes.ZIPCODE=topzipcodes.ZIPCODE.apply(int)
#topzipcodes.CAMIS.apply(int)
topzipcodes.CAMIS=topzipcodes.CAMIS.apply(int)
topzipcodes
```

```python
l=[]
for i in range (len(topzipcodes)):
    t=tuple(topzipcodes.iloc[i])
    l.append(t)

for i in range (len(l)):
    x=list(l[i])
    zipcode=str(int(x[0]))
    camis=int(x[3])
    x[0]=zipcode
    x[3]=camis
    l[i]=tuple(x)
l
    
```
