```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("WebExtract.txt")
```

```python
df
```

```python
df['SD'] = df['SE'] = df.SCORE
```

```python
df.INSPDATE=pd.to_datetime(df.INSPDATE)
df.GRADEDATE=pd.to_datetime(df.GRADEDATE)
df.RECORDDATE=pd.to_datetime(df.RECORDDATE)
c=df.groupby("CAMIS").apply(lambda x: x.sort_values(by='INSPDATE',ascending=False).head(1))
c
```

```python
#Dataframe sorted by 
def SE(x):
    n=np.sqrt(x.count())
    if n==0:
        return None
    ans=np.std(x)
    return(ans)
a=c.groupby('ZIPCODE').agg({"CAMIS":lambda x: x.nunique(),"SCORE":pd.Series.mean,"SD":pd.Series.std,"SE":pd.Series.sem})

a
#b=a.CAMIS.nunique().sort_values(ascending=False).head(100)
```

```python
topzipcodes=a[a.CAMIS>100].sort_values('SCORE',ascending=False)
topzipcodes["ZIPCODE"]=topzipcodes.index
topzipcodes.shape
topzipcodes.head(5)
#topzipcodes[1]

#a=pd.DataFrame({a:topzipcodes.ZIPCODE,b:topzipcodes.SD})
topzipcodes= topzipcodes[['ZIPCODE', 'SCORE', 'SE', 'CAMIS']]
topzipcodes.ZIPCODE=topzipcodes.ZIPCODE.apply(int)
#topzipcodes.CAMIS.apply(int)
topzipcodes.CAMIS=topzipcodes.CAMIS.apply(int)
topzipcodes.shape
topzipcodes
```

```python

a=[(str(int(row[1].ZIPCODE)),row[1].SCORE,row[1].SE,int(row[1].CAMIS)) for row in topzipcodes.iterrows()]
    #print (row[1].ZIPCODE,row[1].SCORE,row[1].SE,row[1].CAMIS)
a
```
