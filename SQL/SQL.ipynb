```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("WebExtract.txt")
```

```python
df['SD'] = df['SE'] = df.SCORE
```

```python
df
```

```python
#Dataframe sorted by 
def SE(x):
    n=np.sqrt(x.count())
    if n==0:
        return None
    ans=np.std(x)
    return(ans)
a=df.groupby('ZIPCODE').agg({"CAMIS":lambda x: x.nunique(),"SCORE":np.mean,"SD":np.std,"SE":SE})
a
#b=a.CAMIS.nunique().sort_values(ascending=False).head(100)
```

```python
topzipcodes=a[a.CAMIS>100].sort_values('CAMIS',ascending=False)
topzipcodes["ZIPCODE"]=topzipcodes.index
topzipcodes.shape
topzipcodes.head(5)
#topzipcodes[1]

#a=pd.DataFrame({a:topzipcodes.ZIPCODE,b:topzipcodes.SD})
topzipcodes= topzipcodes[['ZIPCODE', 'SCORE', 'SE', 'CAMIS']]
topzipcodes
```
