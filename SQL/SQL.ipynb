```python
#c_joined=pd.DataFrame([c_joined.VIOLCODE,c_joined.CUISINECODE])
A=c_joined[['VIOLCODE','CODEDESC']]
#A=A.dropna()
cuisine_list=A.CODEDESC.unique()
violation_list=A.VIOLCODE.unique()
vv=A.groupby("VIOLCODE").count()
A
vv
```

```python
D=A.groupby(['CODEDESC','VIOLCODE']).CODEDESC.count()
D
```

```python
ans=[]
for c1 in cuisine_list:
    for v1 in violation_list:
        #print v1
        if v1 in D[c1].index:
            #print A[v1][c1]
            if D[c1][v1]>=100:
                count=float(D[c1][v1])
                s=float(D[c1].sum())
                p_cond=count/s
                p=vv.loc[v1][0]/float(vv.CODEDESC.sum())
                a=((c1,v1),p_cond/p,count)
                ans.append(a)
            #print D[v1][c1]
            #print p_cond
            #print v1,c1
g=sorted(ans, key=lambda x: -x[1])[:20]
g
```
